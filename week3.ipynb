```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([[1, 2], [3, 4]])

print(arr1)
print(arr2)
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[1:4])   # slicing
print(arr[2])     # indexing
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A + B)
print(np.dot(A, B))
```

```python
A = np.array([[1, 2], [3, 4]])

print(A.T)
print(np.linalg.inv(A))
```

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.mean(arr))
print(np.median(arr))
print(np.var(arr))
print(np.std(arr))
```

```python
rand = np.random.rand(10)

print(rand)
print(np.mean(rand))
print(np.std(rand))
```

```python
A = np.array([[2, 1], [1, -1]])
B = np.array([4, 1])

solution = np.linalg.solve(A, B)
print(solution)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))
print(np.cross(a, b))
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.reshape(3, 2))
print(arr.flatten())
```

```python
import time

# Python list
lst = list(range(1000000))
start = time.time()
sum(lst)
print("List time:", time.time() - start)

# NumPy
arr = np.array(lst)
start = time.time()
np.sum(arr)
print("NumPy time:", time.time() - start)
```

```python
import statistics

data = [1, 2, 2, 3, 4]

# Manual mean
mean = sum(data) / len(data)

print(mean)
print(statistics.median(data))
print(statistics.mode(data))
```

```python
from collections import Counter

data = [1, 2, 2, 3, 3, 3, 4]

freq = Counter(data)
print(freq)
```

```python
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.hist(data)
plt.title("Histogram")
plt.show()
```

```python
trials = 1000
toss = np.random.choice(['H', 'T'], size=trials)

prob_heads = np.sum(toss == 'H') / trials
print(prob_heads)
```

```python
n = 10
p = 0.5

data = np.random.binomial(n, p, 1000)

plt.hist(data)
plt.title("Binomial Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
mu, sigma = 0, 1

y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2/(2*sigma**2))

plt.plot(x, y)
plt.title("Normal Distribution")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

print(np.corrcoef(x, y))
print(np.cov(x, y))
```

```python
data = np.arange(100)

sample = np.random.choice(data, size=10)

print(np.mean(data))
print(np.mean(sample))
```

```python
from scipy import stats

data = [10, 12, 9, 11, 10]

t_stat, p_value = stats.ttest_1samp(data, 10)

print(p_value)
```

```python
from scipy.stats import chi2_contingency

table = [[10, 20], [20, 40]]

chi2, p, dof, expected = chi2_contingency(table)

print(p)
```
