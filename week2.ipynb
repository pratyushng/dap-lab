```python
#4)Swap two numbers
n1 = int(input("Enter the value of the 1st number: "))
n2 = int(input("Enter the value of the 2nd number: "))
print(f"Original numbers: n1 = {n1} and n2 = {n2}")
temp = n1
n1 = n2
n2 = temp
print(f"Swapped numbers: n1 = {n1} and n2 = {n2}")
```

```python
#5)Convert Celsius to Fahrenheit
celsius = float(input("Enter the temperature(in Celsius): "))
fahrenheit = (celsius * 1.8) + 32
print(f"Temperature in Fahrenheit = {int(fahrenheit)}")
```

```python
#6)Convert kilometres to miles
km = float(input("Enter the distance(in km): "))
miles = km * 0.62
print(f"{km} km = {miles:.2f} miles")
```

```python
#7)Find square and cube of a number
num = int(input("Enter the number: "))
sq = num * num
cube = num * num * num
print(f"Square of {num} = {sq}")
print(f"Cube of {num} = {cube}")
```

```python
#8)Calculate simple interest
p = float(input("Enter the principal amount: "))
r = float(input("Enter the rate of interest: "))
t = float(input("Enter the time: "))
si = (p*r*t) / 100
print(f"Simple Interest = {si:.2f}")
```

```python
#9)Calculate compound interest
p = float(input("Enter the principal amount: "))
r = float(input("Enter the rate of interest: "))
t = float(input("Enter the time: "))
amount = p * (1 + r/100) ** t
ci = amount - p
print(f"Compound Interest = {ci:.2f}")
```

```python
#10)Find area of circle, rectangle and triangle
length = float(input("Enter the length: "))
breadth = float(input("Enter the breadth: "))
rect = length * breadth
base = float(input("Enter the base: "))
height = float(input("Enter the height: "))
tri = 0.5 * base * height
radius = float(input("Enter the radius: "))
cir = 3.14 * radius * radius
print(f"\nArea of Triangle = {tri:.2f}")
print(f"Area of Rectangle = {rect:.2f}")
print(f"Area of Circle = {cir:.2f}")
```

```python
#11)Check even or odd
num = int(input("Enter the number: "))
if (num % 2 == 0):
  print(num, "is an even number")
else:
  print(num, "is an odd number")
```

```python
#12)Find largest of two numbers
n1 = int(input("Enter the value of the 1st number: "))
n2 = int(input("Enter the value of the 2nd number: "))
if (n1 > n2):
  print(f"{n1} is greater than {n2}")
else:
  print(f"{n2} is greater than {n1}")
```

```python
#13)Find largest of three numbers
n1 = int(input("Enter the value of the 1st number: "))
n2 = int(input("Enter the value of the 2nd number: "))
n3 = int(input("Enter the value of the 3rd number: "))
if (n1 > n2 and n1 > n3):
  print(f"{n1} is the largest number")
elif (n2 > n3):
  print(f"{n2} is the largest number")
else:
  print(f"{n3} is the largest number")
```

```python
#14)Check positive, negative or zero
num = int(input("Enter the number: "))
if (num > 0):
  print(f"{num} is positive")
elif (num == 0):
  print(f"{num} is zero")
else:
  print(f"{num} is negative")
```
